```python
import pandas as pd
```

## Importing data using pandas read

```python
df = pd.read_csv("/home/chinmay/SuperstorePerformanceProject/Data/SuperstoreData.csv")
df.head(3)
```

## Data Cleaning (Dropping null values, stripping column names and removing any spaces from column names)

```python
df.dropna(inplace=True)
```

```python
df.columns = df.columns.str.strip().str.lower()
```

```python
df.columns = df.columns.str.replace(' ', '')
```

```python
df.head(3)
```

## Converting orderdate and shipdate to pandas datetime objects

```python
df['orderdate'] = pd.to_datetime(df['orderdate'])
```

```python
df['shipdate'] = pd.to_datetime(df['shipdate'])
```

```python
df.head(3)
```

## Exporting the clean data to a new csv file

```python
df.to_csv("/home/chinmay/SuperstorePerformanceProject/Data/SuperstoreCleanData.csv", index=False)
```
